# Face Detection
In this project, we will apply the cascade classification to detect the face in the image.

We need install OpenCV.


```python
import cv2
from matplotlib import pyplot as plt

# Create the path to image and algorithm dataset
imagePath = "./image.jpg"
cascadeClassifierPath = "./haarcascade_frontalface_alt.xml"

# Create Cascade Classifier object
cascadeClassifier = cv2.CascadeClassifier(cascadeClassifierPath)

# Read the original image
image = cv2.imread(imagePath)
plt.imshow(image, cmap = 'gray', interpolation = 'bicubic')

# Convert to Gray Scale image
grayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Apply Cascade Classifier for face detection to image
detecedFaces = cascadeClassifier.detectMultiScale(grayImage)

# Drawing red rectange to the area of detected faces
for (x,y, width, height) in detecedFaces:
    cv2.rectangle(image, (x, y), (x+width, y+height), (0,0,255), 1)
    
# Display image with matplot
plt.imshow(image, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```
