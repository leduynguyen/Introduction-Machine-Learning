# Principal Component Analysis
Provide the low dimensional presentation of database.

Find linear combination feature that are mutually uncorrelated (principal component analysis). 

Help to keep important features

Eigenvalues ... 

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

# Load digit data
digits = load_digits()

# Extract to features and target
features = digits.data
target = digits.target

# Visualiaztion the digits
n_row, n_col, max_n = 2, 5, 10
fig = plt.figure(figsize=(2.0*n_col, 2.26*n_row))
i = 0

print "Viualize the digits"
while i<max_n and i<digits.images.shape[0]:
    p = fig.add_subplot(n_row, n_col, i+1, xticks=[], yticks=[])
    p.imshow(digits.images[i], cmap=plt.cm.bone, interpolation='nearest')
    p.text(0, -1, str(digits.target[i]))
    i=i+1
plt.show()

# Create the model with 10 components corresponding to number 0 to 9
model = PCA(n_components=10)
X_pca = model.fit_transform(features)

# Visulization the PCA
print "Visualization the PCA"

colors = ['black', 'grey', 'red', 'blue', 'green', 'yellow', 'purple', 'orange', 'cyan', 'white']
for i in xrange(len(colors)):
    px = X_pca[:, 0][target == i]
    py = X_pca[:, 1][target == i]
    plt.scatter(px, py, c=colors[i])
    
plt.legend(digits.target_names)    
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
