# K Means Clustering

Unspervised learning algorithm in data mining.
Automatically divides data into clusters of similar items.

Clustering is different with classification or prediction. The given dataset is unkmown and have not labeled. Therefore the model doesnt exit because no relationship between data and features. The cluster algorithm will create a new dataset and divide them to groups.

## _Lloyd algorithm_
    * Initialize the centroids at random.
    * Decide for every point, what centroid is nearest to them.
    * Caculate the new mean of every distinc clusters.
    * Repeat until convergence
    
# Example

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs

# Generate the sample dataset for cluster
X, y = make_blobs(n_samples=100, centers=6, random_state=0, cluster_std=0.6)
print "Plot the sample dataset without cluster"
# Plot the sample dataset
plt.scatter(X[:,0], X[:,1], s=20)
plt.show()

# Create the estimator by using K-means with 6 cluster
estimator = KMeans(6)
estimator.fit(X)
y_means = estimator.predict(X)

# Plot the result of estimator
print "Plot the data after runing the K Means Clustering"
plt.scatter(X[:,0], X[:,1], c=y_means, cmap='rainbow')
plt.show()
```
